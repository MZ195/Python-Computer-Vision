### Detecting faces with CNN (Convolutional Neural Networks)

```python
from PIL import Image
from IPython.display import display
import cv2
import dlib
```

```python
image = cv2.imread('../material/Images/people2.jpg')
cnn_detector = dlib.cnn_face_detection_model_v1('../material/Weights/mmod_human_face_detector.dat')
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
detections = cnn_detector(image, 1)
for face in detections:
    l, t, r, b, c = face.rect.left(), face.rect.top(), face.rect.right(), face.rect.bottom(), face.confidence
    print(c)
    cv2.rectangle(image_rgb, (l, t), (r, b), (255, 255, 0), 2)

display(Image.fromarray(image_rgb))
```
